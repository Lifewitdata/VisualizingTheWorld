```python
# --- WRITE YOUR CODE FOR MODULE 1 TASK 2 ---
cleaned_df = df.dropna()
#--- Inspect data ---
cleaned_df
```

```python
#--- Import matplotlib ---
import matplotlib.pyplot as plt

# --- WRITE YOUR CODE FOR MODULE 1 TASK 3 ---

# Assuming 'cleaned_df' is your DataFrame and it has a 'rating' column
# If the column name is different, replace 'rating' with the correct name

# Count the occurrences of each rating
ratings = cleaned_df['Rating'].value_counts().sort_index()

# Create labels for the pie chart
labels = ratings.index

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(ratings.values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Flipkart Product Ratings')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Add a legend
plt.legend(title="Ratings", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Show the plot
plt.tight_layout()
plt.show()

#--- Inspect data ---
print("Distribution of ratings:")
print(ratings)

# Calculate some statistics
total_ratings = ratings.sum()
average_rating = (ratings * ratings.index).sum() / total_ratings

print(f"\nTotal number of ratings: {total_ratings}")
print(f"Average rating: {average_rating:.2f}")

# Find the most common rating
most_common_rating = ratings.idxmax()
print(f"Most common rating: {most_common_rating} (occurs {ratings[most_common_rating]} times)")

# Calculate the percentage of positive ratings (assuming ratings 4 and 5 are positive)
positive_ratings = ratings[ratings.index >= 4].sum()
positive_percentage = (positive_ratings / total_ratings) * 100

print(f"Percentage of positive ratings (4 and 5): {positive_percentage:.2f}%")
```

```python
#--- Import nltk ---
import nltk
import pandas as pd

# Utilising the NLTK library, download the VADER lexicon.
nltk.download('vader_lexicon')

#--- Import SentimentIntensityAnalyzer from nltk.sentiment.vader ---
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# --- WRITE YOUR CODE FOR MODULE 1 TASK 4 ---

# Initialize the VADER sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Assuming 'cleaned_df' is your DataFrame and it has a 'review_text' column
# If the column name is different, replace 'review_text' with the correct name

# Function to get sentiment scores
def get_sentiment_scores(text):
    return sia.polarity_scores(text)

# Apply sentiment analysis to each review
cleaned_df['sentiment_scores'] = cleaned_df['Review'].apply(get_sentiment_scores)

# Extract individual sentiment scores
cleaned_df['negative'] = cleaned_df['sentiment_scores'].apply(lambda x: x['neg'])
cleaned_df['neutral'] = cleaned_df['sentiment_scores'].apply(lambda x: x['neu'])
cleaned_df['positive'] = cleaned_df['sentiment_scores'].apply(lambda x: x['pos'])
cleaned_df['compound'] = cleaned_df['sentiment_scores'].apply(lambda x: x['compound'])

# Classify sentiment based on compound score
def classify_sentiment(compound):
    if compound >= 0.05:
        return 'Positive'
    elif compound <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

cleaned_df['sentiment'] = cleaned_df['compound'].apply(classify_sentiment)

# Store the results in the 'data' variable
data = cleaned_df

#--- Inspect data ---
print("First few rows with sentiment scores:")
print(data[['Review', 'negative', 'neutral', 'positive', 'compound', 'sentiment']].head())

print("\nDistribution of sentiments:")
print(data['sentiment'].value_counts(normalize=True))

print("\nAverage sentiment scores:")
print(data[['negative', 'neutral', 'positive', 'compound']].mean())

# Correlation between sentiment and rating (if 'rating' column exists)
if 'rating' in data.columns:
    correlation = data['compound'].corr(data['rating'])
    print(f"\nCorrelation between compound sentiment score and rating: {correlation:.2f}")

# Most positive and negative reviews
most_positive = data.loc[data['compound'].idxmax()]
most_negative = data.loc[data['compound'].idxmin()]

print("\nMost positive review:")
print(most_positive['Review'])
print(f"Compound score: {most_positive['compound']}")

print("\nMost negative review:")
print(most_negative['Review'])
print(f"Compound score: {most_negative['compound']}")
```

```python
# --- WRITE YOUR CODE FOR MODULE 1 TASK 5 ---

# Assuming 'data' is the DataFrame from the previous task
# with columns 'positive_score', 'negative_score', and 'neutral_score'

# Sum up positive scores
x = data['positive'].sum()

# Sum up negative scores
y = data['negative'].sum()

# Sum up neutral scores
z = data['neutral'].sum()

#--- Inspect data ---
print("Sum of positive scores:", x)
print("Sum of negative scores:", y)
print("Sum of neutral scores:", z)

# Calculate total score
total_score = x + y + z

# Calculate percentages
positive_percentage = (x / total_score) * 100
negative_percentage = (y / total_score) * 100
neutral_percentage = (z / total_score) * 100

print("\nPercentage breakdown:")
print(f"Positive: {positive_percentage:.2f}%")
print(f"Negative: {negative_percentage:.2f}%")
print(f"Neutral: {neutral_percentage:.2f}%")

# Calculate average scores per review
avg_positive = data['positive'].mean()
avg_negative = data['negative'].mean()
avg_neutral = data['neutral'].mean()

print("\nAverage scores per review:")
print(f"Average positive score: {avg_positive:.4f}")
print(f"Average negative score: {avg_negative:.4f}")
print(f"Average neutral score: {avg_neutral:.4f}")

# Identify reviews with highest positive and negative scores
max_positive_review = data.loc[data['positive'].idxmax()]
max_negative_review = data.loc[data['negative'].idxmax()]

print("\nReview with highest positive score:")
print(f"Score: {max_positive_review['positive']:.4f}")
print(f"Text: {max_positive_review['Review']}")

print("\nReview with highest negative score:")
print(f"Score: {max_negative_review['negative']:.4f}")
print(f"Text: {max_negative_review['Review']}")

# Distribution of compound scores
print("\nDistribution of compound scores:")
print(data['compound'].describe())

# Visualize the distribution of compound scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(data['compound'], bins=20, edgecolor='black')
plt.title('Distribution of Compound Sentiment Scores')
plt.xlabel('Compound Score')
plt.ylabel('Frequency')
plt.show()
```
